# Import required liberaries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px

import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("TSLA.csv")
df
```

# Data Preprocessing

```python
# rename the data
df = df.rename(columns={'Date': 'date', 'Open': 'open', 'High': 'high', 'Low': 'low', 'Close': 'close', 'Adj Close': 'adj_close', 'Volume': 'volume'})
df.head()
```

# change the date into proper formate

```python
df['date'] = pd.to_datetime(df.date)
df.head()
```

# Using the date show the plot 

```python
fig = px.line(df, x='date', y='close', title='Tsla Close price')
fig.show()
```

```python
closedf = df[['date', 'close']]
closedf
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc = MinMaxScaler(feature_range=(0,1))
closedf = sc.fit_transform(np.array(closedf['close']).reshape(-1,1))
closedf
```

# making Training and testing the sets

```python
# create a training and testing size variable
training_size = int(len(closedf)*0.70)   # 70% of the data we have for training purpose
test_size = len(closedf) -training_size

# create a variable Train data and test_data
train_data, test_data = closedf[0:training_size, :], closedf[training_size:len(closedf), :1]
```

```python
train_data
```

```python
test_data
```

# Add some time fector

```python
def create_dataset(dataset, time_step=1):
    # Create a emply list to add the data
    dataX, dataY = [], []
    for i in range(len(dataset)-time_step-1):
        a = dataset[i: (i+time_step), 0] # i = 0,1,2,3,4,5,....,99,100
        # fill the data into the emply list
        dataX.append(a)
        dataY.append(dataset[i + time_step, 0])
    return np.array(dataX), np.array(dataY)
```

# set the time_step

```python
time_step = 50
x_train, y_train = create_dataset(train_data, time_step)
x_test, y_test = create_dataset(test_data, time_step)

print('x_train ', x_train.shape)
print('y_train ', y_train.shape)
print('x_test', x_test.shape)
print('y_test', y_test.shape)
```

```python
from xgboost import XGBRegressor
model = XGBRegressor(n_estimators = 1000)
model.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)
print(pred)
```
